Using value_counts() for quick insights:

For binary or boolean columns like survived or alone, value_counts() immediately tells you the class balance (e.g., how many True vs False).

For categorical columns like sex, embarked, or pclass (which is integer but represents categories 1,2,3), it gives the distribution of categories. For pclass, we might see something like 216 in 1st class, 184 in 2nd, 491 in 3rd (just as an example).

Even for high-cardinality categories (like if we had passenger names, each might be unique), value_counts() could be used to find the most common names or duplicates.

```python
print(df.isnull().sum())
```

**Visualising missing data**

Sometimes it’s helpful to visualize the pattern of missingness. One simple way is to use a heatmap of the boolean mask:



```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

This is a heatmap where each cell is colored if data is present and another color if data is missing. With 891 rows, the heatmap will be quite tall, but the idea is that you’d see vertical stripes for columns with many missing values.

For the Titanic data, a heatmap of missing values would show a big stripe in the deck column (since most values are missing there) and some scattered missing in age (since those 177 missing ages are spread throughout the rows).

It’s not crucial for a small dataset, but with larger data, this can immediately show if there are patterns (e.g., maybe missing values occur mostly in early rows vs later, or certain combinations of missing happen together). In our case, it confirms what we found: the 'Age' and 'Cabin/deck' columns have significant missing data​, whereas others are mostly complete. This visualization is essentially another way to convey what isnull().sum() tells us, but sometimes patterns (like if two columns always have missing values together) can be spotted visually​.

**Checking for duplicate rows**:

A duplicate row means all values in that row match another row in the DataFrame. To find any duplicates:

```python
duplicate_mask = df.duplicated()
print(duplicate_mask.sum())
```

```python
#can also use df.duplicate().any()
df[duplicate_mask].sort_values(by=["Fare"])
```

```python
df.drop_duplicates()
```

**Why these are useful:** Missing values and duplicates are part of **data cleaning checks:*

- Knowing which columns have missing values (and how many) will influence how you clean or model the data. For instance, because deck is mostly missing, you might decide to drop that column entirely or treat “missing deck” as its own category (since “no cabin info” might itself carry some information about those passengers).

- The presence of missing values might also guide EDA: for example, you might wonder if missing ages occur more in certain passenger groups (maybe children? or maybe third class where data collection was poorer?). You could then explore that.

- Duplicates can skew analyses (counting things twice), so it’s good to eliminate them early or at least be aware. Even if none are found, you’ve documented that check.

So, isnull() helps you **identify where data is incomplete**, and duplicated() checks if any data is **redundantly present**. Both are important for ensuring the quality of your dataset before proceeding to deeper analysis or modeling.



**Visual Exploration with Seaborn’s** pairplot

A pair plot (scatterplot matrix) creates a grid of plots such that each numeric variable in the dataset is plotted against each other — giving scatter plots for each pair of variables, and often histograms (or KDEs) on the diagonal to show distributions. It provides a **visual summary of the data**, allowing you to see pairwise relationships at a glance​.

Let’s say we want to visualize how variables like age, fare, and number of relatives relate, possibly colored by survival status. We can do:



```python
sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']], hue='Survived')
plt.show()
```

This command will produce a matrix of plots (roughly 4x4 grid in this case, since we have four numeric variables plus the diagonal). Each cell [i, j] in the grid is a scatter plot of variable i vs variable j, and the diagonal cells [i, i] show the distribution of variable i (often as a histogram or KDE).

By specifying hue='survived', we tell Seaborn to color the points by survival status (e.g., perhaps blue for survived and orange for did not survive). This way, we can visually inspect differences between survivors and non-survivors across these variables.



**Visualizing Missing Data with a Heatmap**

A heatmap of df.isnull() will have the DataFrame’s index on one axis and columns on the other, and a colored cell indicating missing or not. In Python code:

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(),
            yticklabels=False, cbar=False, cmap='viridis')
plt.title("Missing Values in Titanic Dataset")
plt.show()
```
